## BAYES' THEOREM

Bayes' theorem menawarkan formula dalam menghitung nilai probability dari suatu event dengan memanfaatkan pengetahuan sebelumnya dari kondisi terkait atau bisa juga dinamakan Conditional Probability.

## MENGAPA DISEBUT NAIVE ?

## DATASET : BREAST CANCER WISCONSIN (DIAGNOSTIC)

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()

print(f'Dataset = {data.DESCR}')
```

```python
data.keys()
```

```python
X, y = load_breast_cancer(return_X_y = True)
X.shape
```

## TRAINING & TESTING SET

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.2,
                                                    random_state = 0)

print(f'X_train shape = {X_train.shape}')
print(f'X-test shape = {X_test.shape}')
```

## NAIVE BAYES MODEL

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print(f'Accuracy = {acc}')
```

```python
model.score(X_test, y_test)
```
